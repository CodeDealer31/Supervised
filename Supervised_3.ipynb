```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
def regression(X,Y,k):
    
    real_X = np.vstack(X for x in range(k))
    len_X = len(X)
    for i in range(len_X): #each X
        for kk in range(k):#each dim
            real_X[kk, i-1] = (np.sin((kk+1)*np.pi*X[i-1]))
    
    XTX = real_X.dot(real_X.T)
    [ll,oo] = np.linalg.eig(XTX)
    XTY = real_X.dot(Y)
    w = np.linalg.pinv(XTX).dot(XTY)
#     w = np.mat(real_X).I.T.dot(np.mat(Y).T)
    return w
    print(XTX)
    print(XTY)
    print(w)
```

```python
def evaluate_reg(w,X):
    k = len(w)
    real_X = np.vstack(X for x in range(k))
    len_X = len(X)
    for i in range(len_X):
        for kk in range(k):
            real_X[kk, i-1] = (np.sin((kk+1)*np.pi*X[i-1]))
    Y_e = (real_X.T).dot(w)
    return Y_e
```

```python
 def MSE(Y_e,Y):
        MSEvector = [(Y_e[i]-Y[i])**2 for i in range(len(Y))]
        MSE = sum(MSEvector)/len(Y)
        return MSE
```

```python
def generatedata(sample_number=30):
    sampleNo = sample_number
    mu = 0
    sigma = 0.07
    s = np.random.normal(mu, sigma, sampleNo)

    extent = np.linspace(0,1,1000)
    samplex = random.sample(list(extent),sampleNo)  # b2 is out sample point in x

    sample_label = [(math.sin(2*math.pi*i))**2 for i in samplex]  # this is the y value of out sample point
    sample_label = sample_label+s   # add some errors

    return samplex,sample_label
```

```python

[rx,ry] = generatedata(sample_number=30)
dim = [2,5,10,14,18]

x_axix = np.linspace(0,1,100)#for plotting
y_axix = [(math.sin(2*math.pi*i))**2 for i in x_axix]
plt.figure()
plt.plot(rx,ry,'o')
plt.plot(x_axix,y_axix)
plt.title("Orginal plot")
plt.ylabel('y')
plt.xlabel('x')
plt.xlim([0,1])
plt.ylim([0,1])
    

for k in dim:
    x_axix = np.linspace(0,1,100)
    y = np.linspace(0,0,100)
    w = regression(rx,ry,k)
    Y_e = evaluate_reg(w,rx)
    MSEv = MSE(Y_e,ry)
    for p in range(len(x_axix)):
        for kk in range(k):
               y[p] = y[p] + ((np.sin((kk+1)*np.pi*x_axix[p]))) * w[kk]
    plt.figure()
    plt.plot(rx,ry,'o')
    plt.plot(x_axix,y,label = "k = "+str(k)+" MSE="+str(MSEv))
    plt.legend()
    plt.ylabel('y')
    plt.xlabel('x')
    plt.xlim([0,1])
    plt.ylim([0,1])
    

plt.show()
```

```python
tr_mse = []
tt_mse = []
for k in range(19):
    MSE1 = 0
    MSE2 = 0
    for pp in range(100):
        [rx,ry] = generatedata(sample_number=30)  # train_set
        [ttx,tty] = generatedata(sample_number=1000)# test_set
        w = regression(rx,ry,k+1)
        Y_e = evaluate_reg(w,rx)
        tt_e = evaluate_reg(w,ttx)
        MSE1 = MSE1+MSE(Y_e,ry)
        MSE2 = MSE2+MSE(tt_e,tty)
    MSE1 = MSE1/100
    MSE2 = MSE2/100
    tr_mse.append(MSE1)
    tt_mse.append(MSE2)
logmsetr = [math.log(i) for i in tr_mse]
logmsett = [math.log(i) for i in tt_mse]
xaxix = range(19)
plt.figure()
plt.plot(xaxix, logmsetr, 'r*-')
plt.xlabel("k")
plt.ylabel("logMSE")
plt.figure()
plt.plot(xaxix, logmsett, 'r*-')
plt.xlabel("k")
plt.ylabel("logMSE")
plt.show()
```
